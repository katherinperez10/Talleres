```python
CIIU = pd.read_excel("/content/drive/MyDrive/GEIH 2025/Estructura-detallada-CIIU-4AC-2022.xlsx", header=1)
CIIU
```

```python
CIIU = CIIU[["Clase", "Descripción"]]
```

```python
CIIU.shape
```

```python
df_GEIH_2025["RAMA4D_R4"] = df_GEIH_2025["RAMA4D_R4"].astype(float)
CIIU['Clase']= CIIU['Clase'].astype(float)
```

```python
df_GEIH_2025 = df_GEIH_2025.merge(
    CIIU,
    left_on="RAMA4D_R4",
    right_on="Clase",
    how="left")
```

```python
df_GEIH_2025 = df_GEIH_2025.drop(columns=["Clase"])
```
